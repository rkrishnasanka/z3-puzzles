```python
from z3 import *
```

```python
blue, grey, red = Ints('blue grey red')
s = Solver()
s.add(blue > -1, blue < 10)
s.add(grey > -1, grey < 10)
s.add(red > -1, red < 10)
distinct_c = Distinct(blue, grey, red)
sum_c = 100*red + 10*red + red == 3*(100*blue + 10*grey + red)
s.add(distinct_c)
s.add(sum_c)
```

```python
s.check()
```

```python
m = s.model()
for d in m.decls():
    print('%s = %s'%(d.name(), m[d]))
```
